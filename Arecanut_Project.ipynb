**Image Size Distribution:**

```python
import os
import matplotlib.pyplot as plt


dataset_path = '/content/drive/My Drive/Dataset_Images'

# Initialize lists to store image dimensions
widths = []
heights = []

# Iterate through the images and collect dimensions
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(('.jpg', '.png', '.jpeg')):
            image_path = os.path.join(root, file)
            img = plt.imread(image_path)
            height, width, _ = img.shape
            widths.append(width)
            heights.append(height)

# Create histograms for image width and height
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(widths, bins=50, color='blue', alpha=0.7)
plt.xlabel('Image Width')
plt.ylabel('Frequency')
plt.title('Image Width Distribution')

plt.subplot(1, 2, 2)
plt.hist(heights, bins=50, color='green', alpha=0.7)
plt.xlabel('Image Height')
plt.ylabel('Frequency')
plt.title('Image Height Distribution')

plt.tight_layout()
plt.show()
```

```python
import os


file_path = '/content/drive/My Drive/Dataset_Images/High_Quality'

def is_pure_data_file(file_path):
    try:
        with open(file_path, 'rb') as file:
            # Read the first few bytes (adjust the size as needed)
            header = file.read(8)  # Read the first 8 bytes

            if b'PNG' in header or b'JFIF' in header or b'GIF' in header or b'\xff\xd8\xff' in header:
                return False  # It appears to be an image file
            else:
                return True  # It's likely a pure data file
    except Exception as e:
        return False

# Check if the file is a pure data file
if is_pure_data_file(file_path):
    print(f'The file "{file_path}" is likely a pure data file.')
else:
    print(f'The file "{file_path}" appears to be an image file.')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
# Define the path to your dataset directory
dataset_dir = '/content/drive/My Drive/Dataset_Images'

# Define image dimensions (adjust as needed)
image_height, image_width = 128, 128
```

```python
# Initialize an ImageDataGenerator for data augmentation and preprocessing
datagen = keras.preprocessing.image.ImageDataGenerator(
    rescale=1.0/255.0,  # Normalize pixel values to [0, 1]
    rotation_range=20,  # Random rotation
    width_shift_range=0.2,  # Random horizontal shift
    height_shift_range=0.2,  # Random vertical shift
    horizontal_flip=True,  # Random horizontal flip
    validation_split=0.2  # Split the data into training and validation
)
```

```python
# Create a flow_from_directory generator for loading and augmenting image data
batch_size = 32
train_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(image_height, image_width),
    batch_size=batch_size,
    class_mode='categorical',  # Adjust class_mode as needed
    subset='training'
)
```

```python
validation_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(image_height, image_width),
    batch_size=batch_size,
    class_mode='categorical',  # Adjust class_mode as needed
    subset='validation'
)
```

```python
# Define a function to create a CNN model
def create_model(optimizer='adam', dropout_rate=0.2):
    model = keras.Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(image_height, image_width, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(dropout_rate),
        Dense(3, activation='softmax')  # Adjust the number of output units based on your classes
    ])
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Create a custom wrapper function for use with GridSearchCV
def build_fn(optimizer='adam', dropout_rate=0.2):
    return create_model(optimizer=optimizer, dropout_rate=dropout_rate)
```
